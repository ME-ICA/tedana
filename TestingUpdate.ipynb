```python
from TestingTools import *
import shutil
import numpy as np
import os
```

```python
def runTedana(sub,ses,task,topOfBidsPath,echoDir,brain_mask_path,tes_path,echo_dof=None,
                      echo_range=[1,122], echoes=None, echo_step=1,mixm=None,tree="tedana_orig",tes_conversion_factor=10**3,
                      echofilePattern="*_echo-*_desc-preproc_bold.nii*",echofileSortType=2, overwrite=False,savedir=None):
    if not overwrite:
        subjectdir=os.path.join(topOfBidsPath,f"sub-{sub}")
        sessiondir=os.path.join(subjectdir,f"ses-{ses}")
        savedir=os.path.join(sessiondir,"func")
        if not os.path.isdir(topOfBidsPath):
            os.mkdir(topOfBidsPath)
        if not os.path.isdir(subjectdir):
            os.mkdir(subjectdir)

    
        assert os.path.isdir(sessiondir)== False,f"{sessiondir} already exists. Abort."
        os.mkdir(sessiondir)
        os.mkdir(savedir)
        
    print("Echo Directory: {0}".format(echoDir))
    print("Save Directory: {0}".format(savedir))
    print("Mask Directory: {0}".format(brain_mask_path))
    print("Echo Step Size: {0}".format(echo_step))
    if echoes:
        print("Echoes: {0}".format(echoes))
    else:
        print("Echo Range: {0}".format(echo_range))

    if echoes: ## not None
        subsampleEchoes(echoDir=echoDir,
                        savedir=savedir,
                        tes=tes_path,
                        brain_mask_path=brain_mask_path, 
                        save_prefix=f"sub-{sub}_ses-{ses}_task-{task}", 
                        echo_dof=echo_dof,
                        echoes=echoes,
                        tes_conversion_factor=tes_conversion_factor,
                        echofilePattern=echofilePattern,
                        echofileSortType=echofileSortType,
                        topOfBidsPath=topOfBidsPath,
                        mixm=mixm,
                        tree=tree,
                        ovewrite=overwrite)
    else:
        subsampleEchoes(echoDir=echoDir,
                        savedir=savedir,
                        tes=tes_path,
                        brain_mask_path=brain_mask_path, 
                        save_prefix=f"sub-{sub}_ses-{ses}_task-{task}", 
                        echo_dof=echo_dof,
                        echo_start=echo_range[0],
                        echo_stop=echo_range[1],
                        echo_step_size=echo_step,
                        tes_conversion_factor=tes_conversion_factor,
                        echofilePattern=echofilePattern,
                        echofileSortType=echofileSortType,
                        topOfBidsPath=topOfBidsPath,
                        mixm=mixm,
                        tree=tree,
                        overwrite=overwrite)


def runUsingMixm(sub,ses,task,topOfBidsPath,echoDir,brain_mask_path,tes_path,mixm,echo_dof=None,
                      echo_range=[1,122], echoes=None, echo_step=1,tree="tedana_orig",tes_conversion_factor=10**3,
                      echofilePattern="*_echo-*_desc-preproc_bold.nii*",echofileSortType=2):
    subjectdir=os.path.join(topOfBidsPath,f"sub-{sub}")
    sessiondir=os.path.join(subjectdir,f"ses-{ses}")
    savedir=os.path.join(sessiondir,"func")
    if not os.path.isdir(topOfBidsPath):
        os.mkdir(topOfBidsPath)
    if not os.path.isdir(subjectdir):
        os.mkdir(subjectdir)

    assert os.path.isdir(sessiondir)== False,f"{sessiondir} already exists. Abort."
    os.mkdir(sessiondir)
    os.mkdir(savedir)

    print("Copying over files. This may take some time.")
    source_folder=os.path.dirname(mixm)
    for filename in os.listdir(source_folder):
        source_path = os.path.join(source_folder, filename)
        destination_path = os.path.join(savedir, filename)
        if os.path.isfile(source_path):
            shutil.copy2(source_path, destination_path)


    runTedana(sub,ses,task,topOfBidsPath,echoDir,brain_mask_path,tes_path,echo_dof=echo_dof,
              echo_range=echo_range, echoes=echoes, echo_step=echo_step,mixm=mixm,tree=tree,
              tes_conversion_factor=tes_conversion_factor,echofilePattern=echofilePattern,
              echofileSortType=echofileSortType, overwrite=True,savedir=savedir)
```
